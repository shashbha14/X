```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from chromadb import Client
from chromadb.config import Settings
from sklearn.metrics import f1_score

# Load your data
data_path = '/mnt/data/data.csv'
df = pd.read_csv(data_path)

# List of subcategories for embeddings
subcategories = [
    "Child Pornography/Child Sexual Abuse Material (CSAM)",
    "Rape/Gang Rape-Sexually Abusive Content",
    "Sale, Publishing and Transmitting Obscene Material/Sexually Explicit Material",
    "Debit/Credit Card Fraud",
    "SIM Swap Fraud",
    "Internet Banking-Related Fraud",
    # (add the rest of the subcategories as in your list)
]

# Step 1: Initialize model and database
model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight model for embeddings
client = Client(Settings())
collection = client.get_or_create_collection("cybercrime_subcategories")

# Step 2: Generate embeddings for subcategories and store in ChromaDB
subcategory_embeddings = model.encode(subcategories)

for name, embedding in zip(subcategories, subcategory_embeddings):
    collection.add({"name": name, "embedding": embedding})

# Step 3: Define classification function
def classify_text(text):
    text_embedding = model.encode([text])[0]
    # Perform similarity search in ChromaDB
    results = collection.search(query_embedding=text_embedding, top_k=1)
    if results:
        return results[0]["name"]
    return "Unknown"

# Step 4: Classify each entry in your dataset
df['predicted_subcategory'] = df['text_column'].apply(classify_text)

# Step 5: Evaluate the classification (if true labels are available)
# Ensure your labels are available in the dataset, e.g., as 'actual_subcategory'
if 'actual_subcategory' in df.columns:
    f1 = f1_score(df['actual_subcategory'], df['predicted_subcategory'], average='weighted')
    print("F1 Score:", f1)
else:
    print("No actual labels available for evaluation.")
```
